```python
X = np.array([
    1.3,8.6,3.5,10.8,9.3,7.1,4.1,7.9,2.2,10.3,3.2,8.0,9.1,9.6,6.6
],dtype=float)

a = 0
b = 11
X=np.insert(X,0,a)
X=np.insert(X,len(X),b)

X.sort()
print("X = %s"%X)
```

```python
plt.step(X,np.linspace(0,1,len(X)),label="Empitical CDF")
display(Math("\overset{\land}{F}(X)"))
plt.gcf().set_size_inches((5,2.5))
plt.legend()


n = len(X)
Y1 = np.linspace(a,b,n)

X1=sc.uniform.cdf(Y1,a,b)

plt.step(Y1,X1,label="Norm[a,b] CDF")
plt.legend()
plt.gcf().set_size_inches((5,2.5))
plt.show()
```

```python
ks=sc.kstest(X,X1)
print("K-S test result (p-value): ",ks.pvalue)
print("maximum difference: ",ks.statistic)
```

```python
X = [1.31,1.18,1.50,1.06,1.01,1.06,1.33,1.80,1.30,1.35]


plt.hist(X,label="empitical PDF",density=True)

a=0.5
b=1
X1=np.linspace(b,5,100)
plt.plot(X1,sc.expon.pdf(X1,b,a),label="theoretical PDF")

plt.legend()
plt.gcf().set_size_inches((5,2.5))
plt.show()

ks=sc.kstest(X,sc.expon.pdf,args=(b,a))

print("K-S test result (p-value): ",ks.pvalue)
print("maximum difference: ",ks.statistic)
```

```python
from pandas import Series

X = [0.0,0.31,0.7,1.4,0.08,1.93,0.79,1.43,1.42,1.69]

X.sort()
display(Math("X_{i:n} = %s,\space i=\overline{1,n}"%X))

q25 = X[int(len(X)*0.25)]
q75 = X[int(len(X)*0.75)]

display(Math("x_{25} = %s, x_{75} = %s"%(q25,q75)))

Q = q75-q25

print("Q = %s"%round(Q,2))

med = Series(X).median()

print("Median: %s"%med)

plt.hist(X,density=True)
plt.gcf().set_size_inches((5,2.5))
plt.show()
```
