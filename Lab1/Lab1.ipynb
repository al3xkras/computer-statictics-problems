```python
import scipy.stats as sc
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import display
```

```python
X1 = sc.beta.rvs(2.5,0.5,size=100)
X2 = sc.cosine.rvs(loc=1,size=150)
X3 = sc.chi2.rvs(15,size=120)
X4 = sc.erlang.rvs(5,0.8,size=200)
X5 = sc.exponweib.rvs(3,5,size=220)
X6 = sc.gamma.rvs(5,1.5,size=180)
X7 = sc.gompertz.rvs(2,size=130)
X8 = sc.lognorm.rvs(loc=-2,s=4,size=115)
X9 = sc.pareto.rvs(3,size=100)
X10 = sc.wald.rvs(loc=3,size=150)
X11 = sc.laplace.rvs(loc=2,size=120)
X12 = sc.maxwell.rvs(loc=2,size=200)
X13 = sc.rayleigh.rvs(loc=4,size=220)
X14 = sc.logistic.rvs(3,size=180)
X15 = sc.halfnorm.rvs(loc=-4,size=130)
```

```python
s="beta cosine chi2 erlang exponweib gamma gompertz lognorm pareto wald laplace maxwell rayleigh logistic halfnorm"

X_arr = [
    X1,X2,X3,
    X4,X5,X6,
    X7,X8,X9,
    X10,X11,X12,
    X13,X14,X15
]
X_dict = dict(zip(s.split(),X_arr))

theor_pdf = [
    lambda x: sc.beta.pdf(x,2.5,0.5),
    lambda x: sc.cosine.pdf(x,loc=1),
    lambda x: sc.chi2.pdf(x,15),
    lambda x: sc.erlang.pdf(x,5,0.8),
    lambda x: sc.exponweib.pdf(x,3,5),
    lambda x: sc.gamma.pdf(x,5,1.5),
    lambda x: sc.gompertz.pdf(x,2),
    lambda x: sc.lognorm.pdf(x,loc=-2,s=4),
    lambda x: sc.pareto.pdf(x,3),
    lambda x: sc.wald.pdf(x,loc=3),
    lambda x: sc.laplace.pdf(x,loc=2),
    lambda x: sc.maxwell.pdf(x,loc=2),
    lambda x: sc.rayleigh.pdf(x,loc=4),
    lambda x: sc.logistic.pdf(x,3),
    lambda x: sc.halfnorm.pdf(x,loc=-4)
]

theor_cdf = [
    lambda x: sc.beta.cdf(x,2.5,0.5),
    lambda x: sc.cosine.cdf(x,loc=1),
    lambda x: sc.chi2.cdf(x,15),
    lambda x: sc.erlang.cdf(x,5,0.8),
    lambda x: sc.exponweib.cdf(x,3,5),
    lambda x: sc.gamma.cdf(x,5,1.5),
    lambda x: sc.gompertz.cdf(x,2),
    lambda x: sc.lognorm.cdf(x,loc=-2,s=4),
    lambda x: sc.pareto.cdf(x,3),
    lambda x: sc.wald.cdf(x,loc=3),
    lambda x: sc.laplace.cdf(x,loc=2),
    lambda x: sc.maxwell.cdf(x,loc=2),
    lambda x: sc.rayleigh.cdf(x,loc=4),
    lambda x: sc.logistic.cdf(x,3),
    lambda x: sc.halfnorm.cdf(x,loc=-4)
]

X_theor = dict(zip(s.split(),theor_pdf))
X_theor_cdf=dict(zip(s.split(),theor_cdf))
```

```python
plt.hist(X1)
plt.gcf().set_size_inches(9,1)
plt.show()
```

```python
for name in X_dict:
    print(name)
    X = X_dict[name]
    mean=X.mean()
    std=X.std()
    x0 = np.quantile(X,0.25)
    x1 = np.quantile(X,0.75)
    iqr = x1-x0
    
    print("mean = %s"%mean)
    print("std = %s"%std)
    
    fig,ax=plt.subplots(2,1,figsize=(5,3))
    a=ax[0]
    a.boxplot(X)
    a.set_title(name)
    
    a=ax[1]
    a.hist(X,density=True,bins=20)
    
    plt.show()
    
    
    print()
    
    
```

```python
for name in X_dict:
    X = X_dict[name].copy()
    X=np.sort(X)
    pdf=X_theor[name]
    cdf=X_theor_cdf[name]
    Y = np.linspace(X.min(),X.max(),len(X))
    
    plt.gcf().set_size_inches(5,2)
    y = np.arange(1, len(X)+1) / len(X)
    plt.step(X,y)
    plt.plot(Y,[cdf(y) for y in Y])
    plt.title(name)
    plt.show()
    print(sc.kstest(X,cdf))
    
```
